## 下载指定区域的snp信息
- chr: 染色体号
- start: 起始位置
- end: 终止位置
- as_dataframe: 是否输出dataframe
- GeneSymbol[optional]: 只用于输出文字的显示
- featch_all[optional]: 是否下载所有数据
- email[optional]: 邮箱
- fields[optional]: 限定获取的snp的列，见：https://docs.myvariant.info/en/latest/doc/data.html#data-sources\

```python
o = myvariant_downloader.download_ChrRange(chr="chr1", start=69000, end=69500, GeneSymbol="test", as_dataframe=True, fetch_all=False, email="xutingfeng@big.ac.cn", fields="dbnsfp")
```

```python
o = myvariant_downloader.download_ChrRange(chr="chr14", start=63761595, end=64226449, GeneSymbol="test", as_dataframe=True, fetch_all=True, email="xutingfeng@big.ac.cn", fields="dbnsfp")
o
```

## 输入一个bed文件进行下载

```python
import pandas as pd 

bed = pd.read_csv("Clinvar_SNP_Gene_full.bed", names=["chr", "start", "end", "GeneSymbol"], sep="\t")
```

调用`download_ChrRange`进行下载

默认传入
- as_dataframe=True
- fetch_all=True

推荐传入email即可

```python
myvariant_downloader.downloadChrRange_MultiProcessing(bed, processes=10, save_path="test/", email="xutingfeng@big.ac.cn", fields="dbnsfp")
```

## 多线程下载

```python
myvariant_downloader.dow(bed, email="xutingfeng@big.ac.cn", threads_num=5, fields="dbnsfp")
```
